```python
cnn_acc = 87.98
vgg_acc = 90.70
resnet_acc = 86.54
dino_acc = 84.78
```

```python
import matplotlib.pyplot as plt

models = ['CNN', 'VGG16', 'ResNet50', 'DINOV2']
accuracies = [cnn_acc, vgg_acc, resnet_acc, dino_acc]

plt.figure()
plt.bar(models, accuracies)

plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Model Accuracy Comparison')

for i, v in enumerate(accuracies):
    plt.text(i, v + 0.5, str(v), ha='center')

plt.show()
```
